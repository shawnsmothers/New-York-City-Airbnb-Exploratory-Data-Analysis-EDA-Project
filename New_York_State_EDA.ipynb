```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the count chart for the 'neighbourhood_group' column
plt.figure(figsize=(10, 6))  # Increased figure size for better readability
sns.set_style("whitegrid")  # Set a grid style for easier interpretation

# Plotting with sorted neighborhood groups based on their count
sns.countplot(data=df_airBNB,
              x='neighbourhood_group',
              order=df_airBNB['neighbourhood_group'].value_counts().index,
              palette='coolwarm')  # Using a gradient color palette for better visualization

# Setting titles and labels
plt.title('Count of Neighbourhood Groups', fontsize=16)
plt.xlabel('Neighbourhood Group', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotating x-axis labels for better readability if there are many groups
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

Manhatan group is the most frequent of the groups wth Staten Island being the least

```python
# Plotting the count chart for bedroom types
plt.figure(figsize=(10, 6))  # Increased figure size for better readability
sns.set_style("whitegrid")  # Set a grid style for easier interpretation

# Plotting with sorted neighborhood groups based on their count
sns.countplot(data=df_airBNB,
              x='bedrooms',
              order=df_airBNB['bedrooms'].value_counts().index,
              palette='coolwarm')  # Using a gradient color palette for better visualization

# Setting titles and labels
plt.title('Count of bedrooms', fontsize=16)
plt.xlabel('bedrooms', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotating x-axis labels for better readability if there are many groups
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

**Multivariate**

```python
corr = df_airBNB[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'latitude', 'longitude', 'beds']]
corr
```

```python
sns.heatmap(data=corr.corr(), annot=True)
plt.figure(figsize=(8, 6))
```

**Notes**

Price and the number of beds have a weak positive correlation. As the price increases, the number of beds also increases.

The number of reviews and reviews per month have a strong positive relationship, which makes sense since as the review count goes up, the reviews per month would also increase.
